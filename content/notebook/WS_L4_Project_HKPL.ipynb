```python
# First we import the selenium and webdriver
!pip install selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
# Configurate the Chrome Driver
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=options)
```

```python
# Navigate to the target URL
url = "https://webcat.hkpl.gov.hk/search/query?theme=WEB"
driver.get(url)
```

```python
# Check if the connection is successful. Can you print the title of the URL? Cross check it with your browser
print(driver.title)
```
