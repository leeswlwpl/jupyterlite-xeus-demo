```python
a = [1,2,3,4,5]
b = [1,2,3,4,5]

a+b
```

NumPy includes a powerful data structure known as an array. It’s a data structure that organizes multiple items. Each item can be of any type (strings, numbers, or even other arrays).

Arrays are most powerful when they are used to store numbers. This is because arrays give us special ways of performing mathematical operations that are both simpler to write and more efficient computationally.

```python
# Example
my_array = np.array([1, 2, 3, 4, 5, 6])

# We can transform a regular list into a NumPy array by using np.array() and saving the value to a new variable:
my_list = [1, 2, 3, 4, 5, 6]
my_array = np.array(my_list)
```

### 2. Operations with NumPy Arrays II

Arrays can also be added to or subtracted from each other in NumPy, assuming the arrays have the same number of elements.

```python
# Example of adding 2 arrays

a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])
a + b
```

### 5. Selecting Elements from a 1-D Array

NumPy allows us to select elements from an array using their indices.

This is identical to the native python list operations.

```python
# Example

a = np.array([5, 2, 7, 0, 11])
```

If we wanted to select the first element in this array, we would call:

```python
a[0]
```

If we wanted to select multiple elements in the array, we can define a range, such as a[1:3], which will select all the elements from a[1] to a[3], including a[1] but excluding a[3].

```python
a[1:3]
```

```python
a[:3]
```

```python
a[-3:]
```

### 6. Selecting Elements from a 2-D Array

Selecting elements from a 2-d array is very similar to selecting them from a 1-d array, we just have two indices to select from. The syntax for selecting from a 2-d array is `arr[row,column]`.

```python
a = np.array([
    [32, 15, 6, 9 , 14],
    [12, 10, 5, 23, 1],
    [2, 16, 13, 40, 37]
])
```

```python
# Select the second element of the last row
a[2,1]
```

```python
# Select the first column
a[:,0]
```

```python
# Select the second row
a[1]
```

```python
# Select the first 3 elements of the first row
a[0,:3]
```

### 7. Higher dimension

Arrays in NumPy are the data structures with high performance which are suitable for mathematical operations. The three levels of arrays nested inside one another represent the three-dimensional array in python, where each level represents one dimension. We make use of the array function in NumPy to create a three-dimensional array with an object as the parameter passed to it.

rgb_arrays.png

An image is a 3-dimension array.  

- 1st Dimension: Row
- 2nd Dimension: Column
- 3rd Dimension: Pixel

Pixels

- `[255, 255, 255]` is a white pixel.
- `[0, 0, 0]` is a black pixel.
- `[12, 32, 34]` defines a pixel with 12/255 red, 32/255 green and 34/255 blue.

**Example 1**

We will inspect an image array.

```python
# Demonstration
image_array = np.array([

    # ROW 1
    [[100, 43, 78],    # ROW 1 COL 1
     [123, 67, 96],    # ROW 1 COL 2
     [147, 96, 115]],  # ROW 1 COL 3

    # ROW 2
    [[122,  67, 96],   # ROW 2 COL 1
     [161, 114, 122],  # ROW 2 COL 2
     [200, 156, 143]], # ROW 2 COL 3

    # ROW 3
    [[148,  95, 113],  # ROW 3 COL 1
     [202, 155, 145],  # ROW 3 COL 2
     [246, 208, 172]], # ROW 3 COL 3

])
```

```python
# Get the first row
image_array[0]

# 3 pixels will be selected - 9 numbers in total
```

```python
# Get the first column
image_array[:, 0]

# 3 pixels will be selected - 9 numbers in total
```

```python
# Get the center pixel
image_array[1, 1]

# 1 pixel will be selected - 3 numbers in total
```
