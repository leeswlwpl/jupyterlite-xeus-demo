### Danger Restaurant is selling some products: "Cup", "Dish", "Fork", "Spoon"

**Initially, there are 10 Cups, 20 Dishes, 5 Forks and 1 Spoons in the stop.**

1. Create a dictionary call ```inventory```, storing the number of each item. It should be in the format "item_name": item_count.

````python
inventory = {
    "Cup": 10,
    "Dish": 20,
    "Fork": 5,
    "Spoon": 1
}

print(inventory)
````

2. Create a function call ```available_item()```, printing out the inventory with the amount remaining.

````python
def available_item():
    print("="*20)
    for item, count in inventory.items():
        print(f"{item}: {count}")
    print("="*10)

available_item()
````

**For customer to buy an item, we need to write a function for them to call.**

3. Write a function call ```buy(item_name, number_of_item)```, update the dictionary and return a string "OK!"

````python
def buy(item_name, quantity):
    inventory[item_name] -= quantity
    return "OK"
````

4. Try calling ```buy("Cup", 2)```, then call ```available_item()``` to see if the inventory is updated.


````python
buy("Cup", 2)
available_item()
````


**We need to do some validation to see if the input is correct.**

5. Modify the function ```buy(item_name, number_of_item)```, in case the customer buy a wrong item, you return `"No such item"`.


````python
def buy(item_name, quantity):
    if item_name in inventory:
        inventory[item_name] -= quantity
        return "OK"
    else:
        return "No such item."
````


6. Try calling ```buy("Noodles", 2)```, see if the function return ```"No such item"```.


````python
buy("Noodles", 2)
````

7. Modify the function ```buy(item_name, number_of_item)```, in case the customer buy too many, you return `"Sorry, we only have {} {}."`.


````python
def buy(item_name, quantity):
    if item_name in inventory:
        if quantity <= inventory[item_name]:
            inventory[item_name] -= quantity
        else:
            return(f"Sorry, we only have {inventory[item_name]} {item_name}")
    else:
        return "No such item."
````


8. Try calling ```buy("Spoon", 3)```, see if the function returns `"Sorry, we only have 1 spoon."`


````python
buy("Spoon", 3)
````

**Challenge!**
9. Modify the function buy so that ```buy("Fork", 6)``` returns `"Sorry, we only have 5 forks."`

````python
def buy(item_name, quantity):
    if item_name in inventory:
        if quantity <= inventory[item_name]:
            inventory[item_name] -= quantity
        else:
            plural = "s" if inventory[item_name] > 1 else ""
            return(f"Sorry, we only have {inventory[item_name]} {item_name}{plural}")
    else:
        return "No such item"
````

````python
buy("Fork", 2)
buy("Spoon", 3)
````
