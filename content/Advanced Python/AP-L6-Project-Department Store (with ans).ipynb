# Project Department Store

```python
# Load pandas and the df here
import pandas as pd
store_df = pd.read_csv("https://drive.google.com/u/1/uc?id=1kbCDXONvf8Bn4kEW1ssmxdJLDHC-oV-E&export=download")
store_df.head()
```

**Exercise 1**

1. What is the first working and last working week for each Dept-Store?
2. Calculate how many weeks each of them operated.
- Hint1: last working week - first working week + 1
- Hint2: Check the data type of `Date` column

```python
store_df['Date_dt'] = pd.to_datetime(store_df['Date'], format="%d/%m/%Y")
first_last_week = store_df.groupby(["Store", "Dept"])['Date_dt'].agg(['min', 'max'])
first_last_week['number_of_weeks'] = (first_last_week['max'] - first_last_week['min'])\
  .apply(
      # This lambda convert timedelta to number of weeks
      lambda length_timedelta: length_timedelta.days//7  + 1
  )
first_last_week
```

**Exercise 2**

Generate per month sales for store 30 department 2.

```python
# Query is another flavor compare with using boolean selector

store_df.query('Store == 30 and Dept == 2')\
  .set_index("Date_dt")\
  .resample("1m", label="right")['Weekly_Sales']\
  .sum()\
  .head()

# Note that the results' label is the end of each month.
```
